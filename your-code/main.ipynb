## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
# your code here
def generate_uniform_random_numbers(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)
```

## Distribución Normal

1. De la misma manera en el reto de Distribución Uniforme, escriba una función que genere números aleatorios distribuidos normalmente.
1. Genera 1.000 números distribuidos normalmente con media de 10 y desviación típica de 1
1. Genera 1.000 números distribuidos normalmente con una media de 10 y una desviación típica de 50
2. Representa gráficamente las distribuciones de los datos generados.

Resultado esperado:



```python
from scipy.stats import norm 
```

```python
# your code here
def generate_normal_random_numbers(mean, std, count):
    return norm.rvs(loc=mean, scale=std, size=count)
```

```python
# your code here
normal1 = generate_normal_random_numbers(10, 1, 1000)
normal2 = generate_normal_random_numbers(10, 50, 1000)
```

```python
# your code here
# Graficar histogramas
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(normal1, bins=30, edgecolor='black')
plt.title('Normal Distribution: mean=10, std=1, count=1000')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(normal2, bins=30, edgecolor='black')
plt.title('Normal Distribution: mean=10, std=50, count=1000')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
# your code here
def generate_exponential_random_numbers(scale, count):
    return np.random.exponential(scale, count)
```

```python
# Generar números aleatorios exponencialmente distribuidos
exponential1 = generate_exponential_random_numbers(10, 10)
exponential2 = generate_exponential_random_numbers(10, 100)
```

```python
# Graficar histogramas
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(exponential1, bins=10, edgecolor='black')
plt.title('Exponential Distribution: mean=10, count=10')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(exponential2, bins=30, edgecolor='black')
plt.title('Exponential Distribution: mean=10, count=100')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
# your code here
from scipy.stats import expon

# Parámetro lambda para la distribución exponencial
lambda_value = 1 / 10
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
prob_less_than_15 = expon.cdf(15, scale=1/lambda_value) * 100
print(f"Probabilidad de que un cliente pase menos de 15 minutos en el banco: {prob_less_than_15:.2f}%")
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
# your code here
prob_more_than_15 = (1 - expon.cdf(15, scale=1/lambda_value)) * 100
print(f"Probabilidad de que un cliente pase más de 15 minutos en el banco: {prob_more_than_15:.2f}%")
```
